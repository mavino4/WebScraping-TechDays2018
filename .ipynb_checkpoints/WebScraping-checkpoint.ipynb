# Aproximandonos a través del historial de la pagina

```python
rutas = []
for year in soup.find("div", id = "by-date").findAll("div"):
    for link in year.findAll("a"):
        rutas.append(link.get("href"))
        print(link.get("href"))
```

### Ahora vamos a buscar dentro de cada uno de los meses de la página

```python
ruta_i = urllib.request.urlopen( site+rutas[-34] )

ruta_i = bs(ruta_i, "html.parser")

for line in ruta_i.find("div", class_ = "list-archive").findAll("li"):
    print(line.find("a").get("href"))
```

```python
def getUrlMonth(site_url, sub_url):
    links = []
    ruta_i = urllib.request.urlopen( site_url+sub_url)
    ruta_i = bs(ruta_i, "html.parser")
    for line in ruta_i.find("div", class_ = "list-archive").findAll("li"):
        links.append(line.find("a").get("href"))
    return links
```

```python
from tqdm import tqdm
links = []
for link in tqdm(rutas[:20]): 
    links += getUrlMonth(site, link)
```

```python
len(links)
```

```python
linksOrd = sorted([j.split("/")[1:] for j in links])
```

# Reunimos los pasos para poder probarlos en conjunto y de manera más abstracta

```python
receta_i = recetas[np.random.randint(0, len(recetas))]
print(receta_i)
Page_i = bs(urllib.request.urlopen(site+ receta_i), "html.parser")
try: 
    for ingrediente in Page_i.find("ul", class_ = "asset-recipe-list").findAll("li"):
        print (ingrediente.text)
    print(Page_i.find("div", class_ = "asset-recipe-steps").text)
    print(Page_i.find("div", class_="asset-recipe-difficulty").text)
    pic_url = Page_i.find("div", class_ = "article-asset-big").find("img").get("src")
    urllib.request.urlretrieve(pic_url, receta_i.split("/")[-1]+".png")
except:
    pass
```

```python
## Definimos una función que directamente nos obtenga los valores 
def getDatos(receta_i):
    print("Obteniendo datos para: \n" + receta_i)
    Page_i = bs(urllib.request.urlopen(site+ receta_i), "html.parser")
    try: 
        ingredientes = []
        for ingrediente in Page_i.find("ul", class_ = "asset-recipe-list").findAll("li"):
            ingredientes.append(ingrediente.text)
        preparacion = Page_i.find("div", class_ = "asset-recipe-steps").text
        dificultad = Page_i.find("div", class_="asset-recipe-difficulty").text
        pic_url = Page_i.find("div", class_ = "article-asset-big").find("img").get("src")
        urllib.request.urlretrieve(pic_url, receta_i.split("/")[-1]+".png")
        print("%%%%%%%% DATOS OBTENIDOS CON ÉXITO %%%%%%%%%%")
        return [receta_i, dificultad, ingredientes, preparacion]
    except:
        print("%%%%%%%% ERROR %%%%%%%%%%")
        pass
    
```

```python
#Verificamos el número de recetas 
len(recetas)
```

```python
Salida = []
for receta in tqdm(recetas[:10]):
    a = getDatos(receta)
    if a:
        Salida.append(a)
```
